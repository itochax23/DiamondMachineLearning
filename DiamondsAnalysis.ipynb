### Color analysis

### Clarity analysis

```python
diamonds_df['clarity'].value_counts()
```

Note about Clarity: best (FL, flawless) to worst (I3, level 3 inclusions) is as follows
FL,IF, VVS1, VVS2, VS1, VS2, SI1, SI2, I1, I2, I3

```python
plt.figure(figsize = (10,6))
sns.countplot(x = diamonds_df['clarity']);
```
